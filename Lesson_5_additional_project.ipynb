## Приводим колонки к одному порядку:

```python
tm_sales_2 = tm_sales_2.reindex(columns=['SUBS_ID', 'FILIAL_ID', 'PROD_ID', 'ACT_DTTM'])
tm_sales_3 = tm_sales_3.reindex(columns=['SUBS_ID', 'FILIAL_ID', 'PROD_ID', 'ACT_DTTM'])
tm_sales_3
```

## Объединяем 3 таблицы:

```python
tm_total = tm_sales_1.append(tm_sales_2).append(tm_sales_3).reset_index()
tm_total
```

## Исправляем id:

```python
tm_total['st'] = tm_total.SUBS_ID.str.startswith('id')
tm_total
```

```python
tm_total['SUBS_ID'] = 'id'*(1-tm_total.st) + tm_total['SUBS_ID']
tm_total
```

```python
tm_total = tm_total.drop(columns = ['st', 'index'])
tm_total
```

```python
tm_total = tm_total.dropna().reset_index()
tm_total
```

```python
tm_total = tm_total.drop(columns='index')
tm_total
```

## Переводим object в datetime:

```python
tm_total.ACT_DTTM = pd.to_datetime(tm_total.ACT_DTTM, format='%d-%m-%Y %H:%M')
```

```python
tm_total.dtypes
```

```python
logs.START_DTTM = pd.to_datetime(logs.START_DTTM, format='%d-%m-%Y %H:%M')
```

```python
logs.END_DTTM = pd.to_datetime(logs.END_DTTM, format='%d-%m-%Y %H:%M')
```

```python
logs.dtypes
```

```python
logs
```

## Убираем failed purchases:

```python
logs['delta'] = logs.END_DTTM - logs.START_DTTM
logs
```

```python
logs = logs.query('delta >= "5m"')
logs
```

## Джойним:

```python
fin = logs.merge(tm_total, how='inner', on=['PROD_ID', 'SUBS_ID'])
fin
```
