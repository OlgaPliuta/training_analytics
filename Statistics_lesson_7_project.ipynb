```python
import pandas as pd
import seaborn as sns
from scipy import stats
import statsmodels.api as sm
import statsmodels.formula.api as smf

%matplotlib inline
```

```python
df = pd.read_csv('cars.csv')
```

```python
df.isna().sum()
```

```python
df['company'] = df.CarName.str.split().str[0]
df
```

```python
df.company = df.company.replace({'maxda' : 'mazda', 
                                 'Nissan' : 'nissan', 
                                 'porcshce' : 'porsche',
                                 'toyouta' : 'toyota',
                                 'vokswagen' : 'volkswagen',
                                 'vw' : 'volkswagen'})
df.company.unique()
```

```python
df.company.nunique()
```

```python
cols = ['price', 'company', 'fueltype', 'aspiration','carbody', 'drivewheel', 'wheelbase', 'carlength','carwidth', 'curbweight', 'enginetype', 'cylindernumber', 'enginesize', 'boreratio','horsepower']
```

```python
df = df.drop(columns = ['car_ID' , 'CarName'])
df
```

```python
df = df[cols]
df
```

```python
df.corr()
```

```python
df_obj = df.select_dtypes(include=['object'])
df_obj.columns
```

```python
df_dummy = pd.get_dummies(data=df[df_obj.columns], drop_first = True)
df_dummy
```

```python
df_num = df.select_dtypes(include=['float64', 'int64'])
df_num
```

```python
dff = pd.concat([df_num, df_dummy], axis=1)
```

```python
dff
```

```python
cols_for_model = dff.columns
cols_for_model
```

## Predict with only horsepower:

```python
stats.linregress(dff.horsepower, dff.price)
```

```python
r2 = 0.8081388225362212**2
r2
```

## Predict with all variables:

```python
X = dff[cols_for_model[1:]]
X
```

```python
#X = sm.add_constant(X)
X = sm.add_constant(X)
X
```

```python
Y = dff[cols[0].split()]
Y
```

```python
model = sm.OLS(Y, X)
```

```python
results = model.fit()
```

```python
print(results.summary())
```
