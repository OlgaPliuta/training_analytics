## Объединяем выборки по месяцам в одну таблицу:

```python
year = pd.concat([
           pd.read_csv("Lesson_8_tables/bikes_q2_sample_apr.csv"), 
           pd.read_csv("Lesson_8_tables/bikes_q2_sample_may.csv"), 
           pd.read_csv("Lesson_8_tables/bikes_q2_sample_jun.csv"), 
           pd.read_csv("Lesson_8_tables/bikes_q3_sample_july.csv"),
            pd.read_csv("Lesson_8_tables/bikes_q3_sample_aug.csv"),
            pd.read_csv("Lesson_8_tables/bikes_q3_sample_sep.csv"),
            pd.read_csv("Lesson_8_tables/bikes_q4_sample_oct.csv"),
            pd.read_csv("Lesson_8_tables/bikes_q4_sample_nov.csv"),
            pd.read_csv("Lesson_8_tables/bikes_q4_sample_dec.csv")
            ])
year
```

```python
year['start_time'] = pd.to_datetime(year['start_time'])
```

```python
year = year.set_index('start_time')
year
```

```python
year_sub = year.groupby('usertype', as_index=False).resample(rule='D').size()
year_sub
```

```python
year_sub = year_sub.T
year_sub
```

```python
year_sub = year_sub.reset_index()
year_sub
```

```python
year_sub.columns
```

```python
year_sub = year_sub.rename(columns= {0 : 'c', 1 : 's'})
year_sub.columns
```

```python
cust = year_sub.query('c > s')
#[year_sub['0'] > year_sub['1']]
cust
```

## Summer only:

```python
bikes = year
```

```python
bikes_summer = bikes.loc['2018-06-01 00:00:00' : '2018-09-01 00:00:00', :]
bikes_summer
```

```python
top_destination = bikes_summer.groupby('to_station_name', as_index=False) \
                    .agg({'trip_id' : 'count'}) \
                    .sort_values('trip_id', ascending=False).reset_index().to_station_name[0]
top_destination
```

```python
bad_day = bikes_summer.query('to_station_name == @top_destination') \
            .resample(rule='D').size() \
            .idxmin().strftime('%Y-%m-%d')
```

```python
bad_day
```

```python
bikes_summer['weekday'] = bikes_summer.index.strftime('%A')
bikes_summer
```

```python
top = bikes_summer.groupby(['weekday', 'to_station_name'], as_index=False).size()
top
```

```python
top[top.weekday == 'Saturday'].sort_values('size', ascending=False)
```

```python
top[(top.weekday != 'Saturday') & (top.weekday != 'Sunday')].sort_values('size', ascending=False)
```
